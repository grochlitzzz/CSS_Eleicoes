```python
sentences_lemmas = []

for tweet in tweetstexts:
    lemmas = nlp(tweet)
    lemmas = ' '.join([x.lemma_ for x in lemmas]) 
    sentences_lemmas.append(lemmas)
    
for sent, lemma in zip(tweetstexts, sentences_lemmas):
    print(f"\"{sent}\" turns into \"{lemma}\"")
```
